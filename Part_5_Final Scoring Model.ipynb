```python
### Getting Dummies for dataset
X=pd.get_dummies(test1)
y=final_data['customer_type']

```

```python
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X, y)
```

```python
((logreg.coef_)[0])
```

```python
logreg.intercept_
```

```python
columns = X.columns
```

```python
mn=list(zip(X.columns,((logreg.coef_)[0])))
```

```python
test = pd.DataFrame(mn)
```

```python
test.rename(columns={0:'Columns',1:'Coefficients'},inplace=True)
```

```python
test1 = test.append({'Columns' : 'Intercept' , 'Coefficients' :   0.69115405} , ignore_index=True)

### 1.72 is Intercept which comes from logreg.intercept_
```

```python
def root_extract(x):
    if len(x)>6:
        return x[:5]
    else:
        return x[:1]
```

```python
test1['Root'] = test1['Columns'].apply(lambda x: root_extract(x))
```

```python
test1
```

```python
min_coef_sum = test1.groupby('Root').min().sum()
```

```python
min_coef_sum =(min_coef_sum).reset_index().iloc[1,1]
```

```python
max_coef_sum = test1.groupby('Root').max().sum()
```

```python
max_coef_sum=(max_coef_sum).reset_index().iloc[1,1]
```

```python
#### OUR custom scoring range

max_score = 850
min_score = 300
```

```python
def scale_value(x):
    new_value = x*((max_score-min_score)/(max_coef_sum-min_coef_sum))
    return new_value
```

```python
test1['Coefficients']=test1['Coefficients'].astype('float')
```

```python
test1['scaled_value']=test1['Coefficients'].apply(lambda x: scale_value(x))
```

```python
test1.info()
```

```python
test1['scaled_value_copy'] = test1['scaled_value']
```

```python
intercept_value_scaled = ((( 0.69115405- min_coef_sum)/(max_coef_sum-min_coef_sum))*(max_score-min_score))+min_score
```

```python
### Intercept scaled value 
intercept_value_scaled
```

```python
test1.iloc[24,4] =intercept_value_scaled
```

#### Function for generating Customer score based on individual attribute scores

```python
def final_score(x):
    """ intrercept: scaled_score_customer[24] 
    other score """
    score = scaled_score_customer[24] + scaled_score_customer[0]*x[1]+ scaled_score_customer[1]*x[2]+ scaled_score_customer[2]*x[3]+ scaled_score_customer[3]*x[4]+ scaled_score_customer[4]*x[5]\
    + scaled_score_customer[5]*x[6]+ scaled_score_customer[6]*x[7]+ scaled_score_customer[7]*x[8]+ scaled_score_customer[8]*x[9]+ scaled_score_customer[9]*x[10]+ scaled_score_customer[10]*x[11]\
    + scaled_score_customer[11]*x[12]+ scaled_score_customer[12]*x[13]+ scaled_score_customer[13]*x[14]+ scaled_score_customer[14]*x[15]+ scaled_score_customer[15]*x[16]\
    + scaled_score_customer[16]*x[17]+ scaled_score_customer[17]*x[18]+ scaled_score_customer[18]*x[19]+ scaled_score_customer[19]*x[20]+ scaled_score_customer[20]*x[21]\
    + scaled_score_customer[21]*x[22]+ scaled_score_customer[22]*x[23]+ scaled_score_customer[23]*x[24]
    
    return score
```

#### Passing complete row in the apply function

### Mapping Customer Rating 

```python
def map_score(x):
    if x<=450:
        return 'Very Poor'
    elif  x>450 and x<=620:
        return 'Fair'
    elif x>620 and x<=700:
        return 'Good'
    elif x>700 and x<=799:
        return 'Very Good'
    else:
        return 'Exceptional'
```

```python
final_data_with_scorecard['rating']= final_data_with_scorecard['cust_score'].map(map_score)
```
