```python
import pandas as pd
```

```python
def xyz(val):
    
    if val > 0 and val <= 3:
        return 'Z'
    elif val > 3 and val < 10:
        return 'Y'
    elif val >= 10:
        return 'X'
```

```python
sales = pd.read_csv('part_1_sales_DATASET_WITH_ABC_nov_27.csv')
sales['invoicedate'] = pd.to_datetime(sales['invoicedate'])
```

```python
filtered_data = sales[(sales['invoicedate'] > '2017-06-01 00:00:00') & (sales['invoicedate'] < '2018-06-01 00:00:00')]
```

```python
filtered_data.head()
```

```python
chk3 = filtered_data.groupby(['code','year','month'], as_index = False).agg({'sellprice': lambda x: x.sum(),'quantity': lambda x:x.sum()})
```

```python
chk3.head(10)
```

```python
 chk3['count_months']= 1
```

```python
chck4 = chk3.groupby(['code'], as_index = False).agg({'sellprice': lambda x: x.sum(),'quantity': lambda x:x.sum(),'count_months' : lambda x : x.sum()})
```

```python
chck4.head()
```

```python
chck4['class_xyz'] = chck4['count_months'].apply( lambda x : xyz(x))
```

```python
chck4.head()
```

```python
new_xyz = chck4.copy()
```

```python
new_xyz = new_xyz[['code','count_months','class_xyz']]
```

## Defining the Inventory class \
 ### Such that 
   ### 'Green':['AX','AY','BX','BY']
   ### 'Yellow':['AZ','BZ']
   ### 'Red':['CX','CY','CZ']

```python
def invent_label(x):
    labels = {'green':['AX','AY'],'yellow':['BX','AZ','BZ','BY'],'red':['CX','CY','CZ']}
    if x in labels['green']:
        return 'green'
    elif x in labels['yellow']:
        return 'yellow'
    else:
        return 'red'
```

```python
try1['invent_class'] = try1['class_type'].apply(lambda x: invent_label(x))
```
